```python
from sklearn.model_selection import train_test_split
x=data[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p','gender_F', 'gender_M', 'ssc_b_Central',
       'ssc_b_Others', 'hsc_b_Central', 'hsc_b_Others', 'hsc_s_Arts', 'hsc_s_Commerce', 'hsc_s_Science',
       'degree_t_Comm&Mgmt', 'degree_t_Others', 'degree_t_Sci&Tech', 'workex_No', 'workex_Yes', 'salary']]
y=data[['status_Not Placed', 'status_Placed']]
```

```python
x=pd.get_dummies(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error 
mean_squared_error(y_test,pred)
```

```python
r2_score(y_test,pred)
```

```python
mean_absolute_error(y_test, pred)
```
